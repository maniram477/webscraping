### Automation using Selenium 

```python
## Importing selenium webdriver
from selenium import webdriver
```

```python
## options used to run selenium in headless mode
options = webdriver.ChromeOptions()
options.add_argument("--headless")

chrome = webdriver.Chrome(options=options)
```

```python
## Opening a website
chrome.get("http://www.google.com")

## Finding Input field element using XPATH
search_bar = chrome.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[1]/div/div[1]/input')

## Feeding Input text
search_bar.send_keys("Headless mode in selenium site:stackoverflow.com")

## Form submission
search_bar.submit()
```

```python
## Introspection using screen capture
screen = chrome.get_screenshot_as_png()
Image(screen, width=600)
```

```python
## Selecting a result section using XPATH 
results = chrome.find_elements_by_xpath('//*[@id="rso"]/div/div/div/div/div/div[@class="r"]/a[1]')

## Parsing all result urls
all_urls = [ x.get_attribute('href') for x in results ]

## Selecting first result
first_result = results[0]
first_result.click()

## Introspectionr
Image(chrome.get_screenshot_as_png(), width=600)
```
